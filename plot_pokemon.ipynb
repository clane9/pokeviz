```python
from glob import glob
from PIL import Image
from matplotlib import pyplot as plt
import numpy as np
```

```python
image_paths = sorted(glob("pokemon/*.png"))
images = np.stack([Image.open(path) for path in image_paths])
print("Images shape:", images.shape)
```

```python
nr = 10
nc = 17
plotw = ploth = 1.0

f, axs = plt.subplots(nr, nc, figsize=(nc * plotw, nr * ploth))
axs = axs.flatten()

for ii, img in enumerate(images):
    plt.sca(axs[ii])
    plt.imshow(img)
    plt.axis("off")

plt.tight_layout(pad=0.2)
plt.show();
```
